```python
import numpy as np
import pandas as pd
```

## Import Dataset

```python
df=pd.read_csv("Countries.csv")
```

```python
df.head()
```

## Data Extraction

### Country with the highest Population

```python
df['Total_Population']=df['rural_population']+df['urban_population']
df['Total_Population']
```

```python
df.head()
```

```python
df[df['Total_Population']==df['Total_Population'].max()]['country']
```

```python
df.nlargest(1,'Total_Population')
```

### Capital with the highest Population

```python
df[df['Total_Population']==df['Total_Population'].max()]['capital_city']
```

### Country with the Least Population

```python
df[df['Total_Population']==df['Total_Population'].min()][['country','Total_Population']]
```

### Capital with the Least Population

```python
df[df['Total_Population']==df['Total_Population'].min()][['country','capital_city','Total_Population']]
```

### Top 5 Countires with Highest Democratic Score

```python
df.nlargest(5,'democracy_score')[['country','democracy_score']]
```

### Total Number of Unique Regions

```python
df['region'].nunique()
```

### Total Number of Countries in Eastern Europe Region

```python
df[df['region']=='Eastern Europe']['country'].nunique()
```

### Political Leader of a Country with 2nd Highest Population

```python
top2=df.nlargest(2,'Total_Population')
top2.iloc[1]['political_leader']
```

### Country with No Political Leader

```python
df[df['political_leader'].isna()]['country']
```

### Country with Highest Population in Afiaca

```python
df[df['continent']=='Africa'].nlargest(1,'Total_Population')
```
